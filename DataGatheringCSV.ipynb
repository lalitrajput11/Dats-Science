***CONVERTERS*** : Big name can be change to the small

```python
def Rename(name):
  if name == 'Royal Challengers Bangalore':
    return "RCB"
  else:
    return name
```

```python
pd.read_csv('/content/IPL Matches 2008-2020.csv',converters={'team1':Rename,'team2' : Rename})
```
