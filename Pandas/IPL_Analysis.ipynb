```python
import matplotlib.pyplot as plt

# Custom colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Calculate the value counts for toss decisions
toss_decision_counts = df1['toss_decision'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 6))
toss_decision_counts.plot(kind='pie',
                          colors=colors,
                          autopct='%1.1f%%',   # Show percentages
                          startangle=140,     # Start angle for the first slice
                          explode=(0.1, 0),   # Explode first slice
                          shadow=True)        # Add a shadow

# Add a title
plt.title('Toss Decision Distribution')

# Add a legend with the actual counts
plt.legend(labels=[f'{label} ({count})' for label, count in zip(toss_decision_counts.index, toss_decision_counts)],
           loc='upper right')

# Show the plot
plt.show()
```
